```
def kmeans(X, k, maxiter,yrr, seed = None):
    """
    specification de nombre de clusters et le nombre maximal des itiration 
    """
    n_row, n_col = X.shape

    # le choix aleatoire de k parmi la data k
    if seed is not None:
        np.random.seed(seed)
    
    rand_indices = np.random.choice(n_row, size = k)
    centroids =yrr

    for itr in range(maxiter):
        # calcul de la distance entre les centroide et les point de data
        # assigement de chaque point au cluster le plus proche
        distances_to_centroids = pairwise_distances(X, centroids, metric = 'euclidean')
        cluster_assignment = np.argmin(distances_to_centroids, axis = 1)

        # selection de tout la data qui contient au cluster  i and compte
        # la nouvelle cluster obtenue
        new_centroids = np.array([X[cluster_assignment == i].mean(axis = 0) for i in range(k)])
        
        #si le centroide est le meme ,
        #l'algorithme converge
        if np.all(centroids == new_centroids):
            break
        
        centroids = new_centroids
    
    return centroids, cluster_assignment
```
